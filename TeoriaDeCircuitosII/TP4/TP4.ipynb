```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctl
import scipy.signal as sig
```

```python
num = [1, 0, -1]
den = [1, 0, 0]
H_z = sig.TransferFunction(num, den,dt=0.01)
n, mag, omega = sig.dbode(H_z)

plt.figure()
plt.plot(n, mag)    # Bode magnitude plot
plt.grid()
plt.figure()
plt.plot(n, omega)  # Bode phase plot
plt.grid()
plt.show()
```
